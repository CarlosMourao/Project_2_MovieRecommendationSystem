```python
# table with Actor and Actress more representative

name = ['Sameera Sherief', 'Delhi Kumar']
category = ['Actress', 'Actor']
value_count = ['10126', '7834']

data = {'Name':name,
        'Category':category,
        'Total':value_count}

top_actor_actress = pd.DataFrame(data)
```

```python
# table for the Top Actor and Actress
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize = (15, 3))
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
table_top_ten_shows= ax.table(cellText=top_actor_actress.values, colLabels=top_actor_actress.columns,
                              loc='center', colColours =["tab:blue"] * 100, colLoc = 'center' ,
                              cellLoc='center', colWidths=[0.5 for x in top_actor_actress.columns],
                              bbox=[-2, 0, 1, 2]) #bbox = [position on left (need to be negative to stay on left), position up/down, lenght of table/row, height of table/columns]
table_top_ten_shows.auto_set_font_size(False)
table_top_ten_shows.set_fontsize(24)
fig.tight_layout()
plt.show()
```

```python
#### VISUALIZATION OF Percentage between actor and actress #####

import matplotlib.pyplot as plt
import seaborn as sns

y = total_actor_actress['category'].value_counts()
labels = ['Actor', 'Actress']

fig = plt.figure(figsize=(15,10))
plt.pie(y, startangle = 155, autopct=lambda p:f'{p:.2f}%', wedgeprops={'linewidth': 0.25, 'edgecolor': 'black'}, textprops={'fontsize': 18})
plt.legend(labels, loc = 'upper left', prop={'size': 18})

plt.show()
```
