# **TOTAL DIRECTORS**

```python
import pandas as pd

directors = pd.read_csv('/content/drive/MyDrive/Projects/Project 2_CINEMA_Movies/new_title_crew_clean.csv', sep = ',') 
```

```python
directors.info(verbose = True, null_counts = True)
```

```python
directors_count = pd.DataFrame(directors['primaryName'].value_counts())
directors_count
```

```python
directors_count_final = directors_count.iloc[:5]
directors_count_final
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize = (12,8))
ax = sns.barplot(data = directors_count_final, x = directors_count_final.index, y = 'primaryName')
bar = plt.bar(directors_count_final.index , directors_count_final['primaryName'])                 
for b in bar:
  yval = b.get_height()
  plt.text(b.get_x()+0.25, yval, yval)
# plt.show()
# plt.bar(x = directors_count.iloc[:5])
plt.show()
```

# **DIRECTORS ON SERIES**

```python
import pandas as pd
##### Episodes #####
series = pd.read_csv('/content/drive/MyDrive/Projects/Project 2_CINEMA_Movies/episodes_cleaned.csv', sep = ',')
```

```python
series
```

```python
#Merge of Directors and Series
directors_series = pd.merge(left = directors,
                            right = series,
                            left_on = 'title_id',
                            right_on = 'title_id')
```

```python
directors_series.head()
```

```python
# variable with the value_counts

directors_series_val = pd.DataFrame(directors_series['primaryName'].value_counts())
directors_series_val
```

```python
# variable with the Top 5 value_counts

directors_series_final = directors_series_val.iloc[:5]
directors_series_final
```

```python
#Visualization of the Top 5 Directors in Series

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize = (12,8))
ax = sns.barplot(data = directors_series_final, x = directors_series_final.index, y = 'primaryName')
bar = plt.bar(directors_series_final.index , directors_series_final['primaryName'])                 
for b in bar:
  yval = b.get_height()
  plt.text(b.get_x()+0.25, yval+100, yval, fontsize = 14)

ax.set_xticklabels(['Nivedita Basu', 'Saibal Banerjee', 'Anil v Kumar', 'Mário Márcio Bandarra', 'Henrique Martins'], size = 12)
ax.set_xlabel("Director's Name", size = 15, labelpad = 15)
ax.set_ylabel("Total Series", size = 15, labelpad = 10)
# plt.xlabel("Diretor's Name")
# plt.ylabel('Value Count')
# plt.show()
# plt.bar(x = directors_count.iloc[:5])
plt.show()
```

```python
##### Save plot ####

plt.savefig('Directors more common.png')
```

# **DIRECTORS ON MOVIES**

```python
##### Movies #####
movies = pd.read_csv('/content/drive/MyDrive/Projects/Project 2_CINEMA_Movies/not_in_series_films.zip', sep = ',')
```

```python
movies
```

```python
##### Merged DIRECTORS and MOVIES ######
directors_movies = pd.merge(left = directors,
                            right = movies,
                            left_on = 'title_id',
                            right_on = 'title_id')
```

```python
directors_movies.head()
```

```python
### value count of directors from Movies

directors_movies_val = pd.DataFrame(directors_movies['primaryName'].value_counts())
directors_movies_val
```

```python
# Top 5 directors from movies

directors_movies_final = directors_movies_val.iloc[:5]
directors_movies_final
```

```python
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize = (12,8))
ax = sns.barplot(data = directors_movies_final, x = directors_movies_final.index, y = 'primaryName')
bar = plt.bar(directors_movies_final.index , directors_movies_final['primaryName'])                 
for b in bar:
  yval = b.get_height()
  plt.text(b.get_x()+0.25, yval+25, yval, size = 15)
# plt.title('Top 5 Directors in Movies')
ax.set_xticklabels(['Bruno De Paola', 'Konatan', 'Kits Fernandez', 'Desireé Markgraaff', 'Martin Dano'], size = 13)
ax.set_xlabel("Director's Name", size = 15, labelpad = 15)
ax.set_ylabel("Total Movies", size = 15, labelpad = 10)
# plt.xlabel("Diretor's Name")
# plt.ylabel('Value Count')
# plt.show()
# plt.bar(x = directors_count.iloc[:5])
plt.show()
```
