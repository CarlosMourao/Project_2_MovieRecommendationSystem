```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
genre_mentions = pd.read_csv('/content/genre_mentions_cleaned.csv')
```

```python
series_on_air = pd.read_csv('/content/series_cleaned.csv')
```

# Genres mentions


```python
genre_mentions_final = genre_mentions.head()                          #If we want top 10 viz, just put 10 inside of head and run this and next cell
genre_mentions_final
```

```python
fig, ax = plt.subplots(figsize = (20, 9))

#ax = sns.histplot(data = genre_mentions_final, x = 'Genres')                    #If time, try, would look really good
ax = sns.barplot(data = genre_mentions_final, x = 'Genres', y = 'Mentions')

bar = plt.bar(genre_mentions_final['Genres'] , genre_mentions_final['Mentions']) 
for b in bar:
  yval = b.get_height()
  plt.text(b.get_x()+0.23, yval+20000, yval, fontsize = 14)

ax.set_xticklabels(['Drama', 'Comedy', 'Short', 'Talk-Show', 'Documentary'], size = 14)
ax.set_xlabel('Genres', size = 15, labelpad = 15)
ax.set_ylabel('Mentions', size = 15, labelpad = 15)
plt.show()
```

# Series on Air

```python
series_on_air_sorted = series_on_air.sort_values(by = 'years_on_air', ascending = False)
series_on_air_final = series_on_air_sorted.head()   
series_on_air_final = series_on_air_final.drop(columns=['series_id'])
series_on_air_final = series_on_air_final[["seriesTitle", "startYear", "endYear","years_on_air"]]
series_on_air_final = series_on_air_final.rename(columns={"seriesTitle":"Title", "startYear":"Start Year","endYear":"End Year","years_on_air":"Years on Air"})                   #SAME AS BEFORE, If we want top 10, put 10 inside head and run!
series_on_air_final.head()
```

```python
fig, ax = plt.subplots(figsize = (12, 3))


fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')

table_top_ten_shows= ax.table(cellText=series_on_air_final.values, colLabels=series_on_air_final.columns, loc='center', colColours =["tab:blue"] * 100, colLoc = 'center' , cellLoc='center', colWidths=[0.5 for x in series_on_air_final.columns],
          bbox=[0., -2, 1.5, 3],  )
table_top_ten_shows.auto_set_font_size(False)
table_top_ten_shows.set_fontsize(14)

fig.tight_layout()

plt.show()
```
